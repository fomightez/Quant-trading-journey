# Trying install and import in a session with the repo content present

The content from the repo is already present in the active session if launched correct.

--------

### Two steps to demonstrate it works

From inside Jupyter use the magic install command that insures it installs to the correct environment.  
The magic pip command variation was added in 2019 to ensure the install occurs in the environment where the kernel is running that backs the active notebook.    
See more about the modern `%pip install` command [here](https://discourse.jupyter.org/t/why-users-can-install-modules-from-pip-but-not-from-conda/10722/4?u=fomightez).

-----

````python
%pip install -e .
````

Let that run and then when it finishes...  
**IMPORTANT. RESTART THE KERNEL!!!**

````python
from quant_utils.data import download_data
````

That works! No error this time. 

You could add `uv` into this mix as others have suggested; however, then you still need to insure the correct environment.
